```python
#Separate Features
df_numdis_features=['bedrooms','floors','bathrooms']
df_numcont_features=['sqft_living','sqft_lot','sqft_above','sqft_living15','sqft_lot15']
df_catbin_features=['waterfront']
df_catdis_features=['condition','grade','view']
df_loc_features=['zipcode','lat','long']
df_dae_features=['date','yr_built','yr_renovated']
```

```python
#drop the data with bedroom as 33
df.drop(df[df['bedrooms']==33].index,inplace=True)
```

```python
#Keep a backup for df
df_backup=df
```

```python
df['bathrooms']=df['bathrooms'].round()
```

```python
df.info()
```

```python
df['sqft_basement']=df['sqft_basement'].replace(0,'NIL')
```

```python
df.head()
```

```python
df[df['sqft_basement']!= 'NIL'].sqft_basement.values
```

```python
df['sqft_basement']=pd.cut(df[df['sqft_basement']!= 'NIL'].sqft_basement,4,labels=['small','medium','high','luxury'])
```

```python
for i in df_numcont_features:
    df[i]=pd.cut(df[i],4,labels=['small','medium','high','luxury'])
```

```python
#g = g.cat.add_categories([1])
#g.fillna(1)
df['sqft_basement']=df['sqft_basement'].cat.add_categories('NIL')
df['sqft_basement'].fillna('NIL',inplace=True)
```

```python
df.head()
```

```python
for i in df_catdis_features:
    df[i]=pd.cut(df[i],4,labels=['poor','good','verygood','excellent'])
```

```python
df.head()
```

```python
df['year_sold']=df.date.str.slice(0,4)
```

```python
df['yr_renovated']=df['yr_renovated'].astype(int)
df['year_sold']=df['year_sold'].astype(int)
```

```python
df['yrafter_renovaton']=df['year_sold']-df[df['yr_renovated']!= 0].yr_renovated
```

```python
df['yrof_house']=df['year_sold']-df['yr_built']
```

```python
df['yrafter_renovaton'].fillna(df['year_sold']-df['yr_built'],inplace=True)
```

```python
df['bathrooms']=df['bathrooms'].astype(int)
```

```python
df['floors']=df['floors'].astype(int)
```

```python
df.head()
```

```python
df.drop(['date','yr_built','yr_renovated','zipcode','lat','long'],axis=1,inplace=True)
```

```python
df['yrafter_renovaton']=df['yrafter_renovaton'].astype(int)
```

```python
df_bkup2=df
```

```python
df.columns
```

```python
df_years=['year_sold', 'yrafter_renovaton', 'yrof_house']
for i in df_years:
    df[i]=pd.cut(df[i],4,labels=['veryold','old','okay','new'])
```

```python
df.head()
```

```python
df_final=pd.get_dummies(df,columns=[ 'sqft_living', 'sqft_lot', 
       'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement',  'sqft_living15',
       'sqft_lot15', 'year_sold', 'yrafter_renovaton', 'yrof_house'],drop_first=True)
```
